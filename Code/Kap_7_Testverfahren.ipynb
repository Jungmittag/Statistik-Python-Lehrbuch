## 7.1 Einstichprobentest für Anteilswerte

```python
from statsmodels.stats.proportion import proportions_ztest

#Eingabeparameter:
x = 60 # Anzahl der zufriedenen Kunden
n = 100 # Gesamtanzahl der Kunden
theta_H0 = 0.7 # Hypothetischer Anteil unter H0

# Durchführung des Z-Tests für Proportionen (zweiseitig)
stat, pval = proportions_ztest(x, n, theta_H0, alternative='two-sided', prop_var=theta_H0)

print(f"Teststatistik (z-Wert): {stat:.4f}")
print(f"p-Wert: {pval:.4f}")

# Interpretation
alpha = 0.05
if pval < alpha:
    print(f"Da der p-Wert ({pval:.4f}) kleiner ist als das Signifikanzniveau alpha ({alpha}), lehnen wir H0 ab.")
    print("Es gibt signifikante Evidenz, dass der Anteil der zufriedenen Kunden nicht 70% beträgt.")
else:
    print(f"Da der p-Wert ({pval:.4f}) größer ist als das Signifikanzniveau alpha ({alpha}), können wir H0 nicht ablehnen.")
    print("Es gibt keine ausreichende Evidenz, um die Behauptung des Herstellers zu widerlegen.")
```

## 7.2 Einstichprobentest für das arithmetische Mittel

```python
import pandas as pd
from scipy import stats

df_Fuellmengen=pd.read_csv('Data/Fuellmengen.csv', delimiter=';')
print(df_Fuellmengen.head(8))

mittelwert_stichprobe = df_Fuellmengen['Fuellmenge'].mean()
std_stichprobe = df_Fuellmengen['Fuellmenge'].std()
n = len(df_Fuellmengen['Fuellmenge'])
print(f"Mittelwert der Stichprobe: {mittelwert_stichprobe:.2f}")
print(f"Standardabweichung der Stichprobe: {std_stichprobe:.2f}")
print(f"Anzahl der Datenpunkte in der Stichprobe: {n}")

# Hypothetischer Mittelwert unter H0
mu_H0 = 1000

# Durchführung des Einstichproben-t-Tests
stat, pval = stats.ttest_1samp(df_Fuellmengen['Fuellmenge'], mu_H0, alternative='two-sided')

print(f"Teststatistik (t-Wert): {stat:.4f}")
print(f"p-Wert: {pval:.4f}")

# Interpretation
alpha = 0.05
if pval < alpha:
    print(f"Da der p-Wert ({pval:.4f}) kleiner ist als das Signifikanzniveau alpha ({alpha}), lehnen wir H0 ab.")
    print("Es gibt signifikante Evidenz, dass die durchschnittliche Füllmenge nicht 1000 ml beträgt.")
else:
    print(f"Da der p-Wert ({pval:.4f}) größer ist als das Signifikanzniveau alpha ({alpha}), können wir H0 nicht ablehnen.")
    print("Es gibt keine ausreichende Evidenz, um zu schlussfolgern, dass die Füllmenge von 1000 ml abweicht.")
```

```python
import numpy as np
from scipy import stats

# --- Gegebene Daten der Stichprobe ---
stichproben_mittelwert = 992.27
stichproben_stdabw = 13.56
stichproben_umfang = 30

# --- Hypothese ---
grundgesamtheit_mittelwert_nullhypothese = 1000  # Der Populationsmittelwert unter der Nullhypothese (μ0)
alpha = 0.05                             # Das Signifikanzniveau

# --- Berechnung der Teststatistik ---

# Freiheitsgrade
df = stichproben_umfang - 1

# Standardfehler des Mittelwerts
standardfehler = stichproben_stdabw / np.sqrt(stichproben_umfang)

# t-Wert berechnen
t_wert = (stichproben_mittelwert - grundgesamtheit_mittelwert_nullhypothese) / standardfehler

# p-Wert berechnen (zweiseitiger Test)
# stats.t.sf ist die Survival Function (1 - CDF)
p_wert = stats.t.sf(np.abs(t_wert), df) * 2

# --- Ergebnisinterpretation ---

# Signifikanz prüfen
ist_signifikant = p_wert < alpha

print(f"--- Ergebnisse des Einstichproben-t-Tests ---")
print(f"Stichprobenmittelwert: {stichproben_mittelwert:.2f}")
print(f"Stichprobenstandardabweichung: {stichproben_stdabw:.2f}")
print(f"Stichprobenumfang: {stichproben_umfang}")
print(f"Populationsmittelwert (Nullhypothese): {grundgesamtheit_mittelwert_nullhypothese:.2f}")
print(f"Freiheitsgrade (df): {df}")
print(f"Berechneter t-Wert: {t_wert:.4f}")
print(f"Berechneter p-Wert: {p_wert:.4f}")
print(f"Signifikanzniveau (alpha): {alpha}")

if ist_signifikant:
    print(f"Ergebnis: Der p-Wert ({p_wert:.4f}) ist kleiner als alpha ({alpha}), daher verwerfen wir die Nullhypothese.")
    print(f"Es gibt einen signifikanten Unterschied zwischen dem Stichprobenmittelwert und dem Populationsmittelwert unter der Nullhypothese.")
else:
    print(f"Ergebnis: Der p-Wert ({p_wert:.4f}) ist größer oder gleich alpha ({alpha}), daher verwerfen wir die Nullhypothese nicht.")
    print(f"Es gibt keinen signifikanten Unterschied zwischen dem Stichprobenmittelwert und dem Populationsmittelwert unter der Nullhypothese.")
```

## 7.3 Einstichprobentest für die Varianz

```python
import numpy as np
from scipy.stats import chi2

# 1. Eingabedaten
# ----------------------------------------------------------------------

# Stichprobe der Stückkosten (€)
kosten_daten = np.array([
    10.25, 9.80, 10.00, 10.15, 9.90,
    10.30, 10.05, 10.10, 9.75, 10.40,
    9.95, 10.30, 10.00, 10.10, 10.25,
    9.90, 10.00, 10.15, 10.05, 10.20,
    9.90, 10.35, 9.85, 10.10, 10.20
])

# Vorgegebene Varianz (unter H0)
sigma2_0 = 0.0225

# Signifikanzniveau
alpha = 0.05

# 2. Berechnung der Teststatistik
# ----------------------------------------------------------------------

# Stichprobengröße (n) und Freiheitsgrade (df = n - 1)
n = len(kosten_daten)
df = n - 1

# Berechnung der Stichprobenvarianz (s^2)
stichproben_varianz = np.var(kosten_daten, ddof=1)

# Berechnung der Chi-Quadrat-Teststatistik (X^2)
# X^2 = (n - 1) * s^2 / sigma_0^2
chi2_teststatistik = (df * stichproben_varianz) / sigma2_0

# 3. Bestimmung des p-Werts (einseitiger Test)
# ----------------------------------------------------------------------

# Wir testen auf HA: sigma^2 > sigma_0^2, daher wird die Wahrscheinlichkeit für die Grenze des rechten Ablehnungsbereichs benötigt
# (Survival Function, sf) der Chi-Quadrat-Verteilung verwendet (sf = 1-cdf)
p_wert = chi2.sf(chi2_teststatistik, df)

# 4. Ergebnis und Schlussfolgerung
# ----------------------------------------------------------------------

print(f"**Analyse der Stückkosten-Varianz**")
print("-" * 40)
print(f"Stichprobengröße (n): {n}")
print(f"Freiheitsgrade (df): {df}")
print(f"Ziel-Varianz (sigma_0^2): {sigma2_0} €^2")
print(f"Signifikanzniveau (alpha): {alpha}")
print("-" * 40)
print(f"Berechnete Stichprobenvarianz (s^2): {stichproben_varianz:.5f} €^2")
print(f"Chi-Quadrat-Teststatistik (X^2): {chi2_teststatistik:.4f}")
print(f"P-Wert (einseitig rechts): {p_wert:.4f}")
print("-" * 40)

# Entscheidungsregel
if p_wert < alpha:
    entscheidung = "Ablehnung der H0"
    schlussfolgerung = (
        "Die Stichprobe liefert signifikante Beweise dafür, dass die wahre Varianz "
        "der Stückkosten **größer** als 0,0225 €^2 ist. Der Prozess ist **instabil** "
        "und die Ursachen für die hohe Kostenschwankung müssen untersucht werden."
    )
else:
    entscheidung = "Nicht-Ablehnung der H0"
    schlussfolgerung = (
        "Es gibt keine ausreichenden statistischen Beweise, um die H0 abzulehnen. "
        "Die Varianz der Stückkosten liegt **innerhalb** des akzeptablen Rahmens (kleiner oder gleich 0,0225 €^2). "
        "Die Prozessstabilität ist gewährleistet."
    )

print(f"Entscheidung bei Signifikanzniveau alpha={alpha}: **{entscheidung}**")
print(f"Betriebswirtschaftliche Schlussfolgerung: {schlussfolgerung}")
```

## 7.4 Zweistichprobentests für Anteilswerte

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# --- 1. Daten definieren ---
# Anzahl der "Erfolge" (Klicks)
clicks_1 = 180
clicks_2 = 235

# Anzahl der Beobachtungen (Empfänger)
n_1 = 2500
n_2 = 2700

# Zusammenstellen der Arrays für die Funktion
# erfolge: Array mit Anzahl der Erfolge [Klicks 1, Klicks 2]
erfolge = np.array([clicks_1, clicks_2])

# nobs: Array mit Anzahl der Beobachtungen (number of observations) [Empfänger 1, Empfänger 2]
nobs = np.array([n_1, n_2])

# --- 2. Berechnung der beobachteten Anteilswerte (zur Info) ---
anteil_1 = clicks_1 / n_1
anteil_2 = clicks_2 / n_2

print(f"Klickrate Variante 1: {anteil_1:.4f} ({anteil_1*100:.2f}%)")
print(f"Klickrate Variante 2: {anteil_2:.4f} ({anteil_2*100:.2f}%)")
print("-" * 30)

# --- 3. Durchführung des Hypothesentests ---
# value=0 bedeutet, wir testen auf eine Differenz von 0
# alternative='two-sided' für den zweiseitigen Test
stat, p_value = proportions_ztest(count=erfolge, nobs=nobs, value=0, alternative='two-sided')

print(f"z-Statistik: {stat:.4f}")
print(f"p-Wert:      {p_value:.4f}")

# --- 4. Entscheidung ---
alpha = 0.05

print("-" * 30)
if p_value < alpha:
    print("Ergebnis: Die Nullhypothese wird abgelehnt.")
    print("Interpretation: Es besteht ein signifikanter Unterschied zwischen den Klickraten.")
else:
    print("Ergebnis: Die Nullhypothese wird nicht abgelehnt.")
    print("Interpretation: Der Unterschied könnte zufällig sein.")
```

## 7.5 Zweistichprobentests für arithmetische Mittel

```python
import pandas as pd
from scipy import stats

df_Fonds=pd.read_csv('Data/Zweistichproben_Fonds.csv', delimiter=';')
print(df_Fonds.head(8))

# 2. Durchführung des Hypothesentests (Welch-Test)
# equal_var=False aktiviert den Welch-Test (Berücksichtigung ungleicher Varianzen)
stat, p_value = stats.ttest_ind(df_Fonds['Fonds_A'], df_Fonds['Fonds_B'], equal_var=False, nan_policy='omit')

print(f"Mittelwert Fonds A: {df_Fonds['Fonds_A'].mean():.2f}% (StdAbw: {df_Fonds['Fonds_A'].std():.2f})")
print(f"Mittelwert Fonds B: {df_Fonds['Fonds_B'].mean():.2f}% (StdAbw: {df_Fonds['Fonds_B'].std():.2f})")
print("-" * 40)
print(f"t-Statistik (Welch): {stat:.4f}")
print(f"p-Wert (zweiseitig): {p_value:.4f}")

# 3. Entscheidung
alpha = 0.05
print("-" * 40)
if p_value < alpha:
    print("Ergebnis: Die Nullhypothese wird abgelehnt.")
    print("Interpretation: Es besteht ein signifikanter Unterschied zwischen den Renditen.")
else:
    print("Ergebnis: Die Nullhypothese wird nicht abgelehnt.")
    print("Interpretation: Der Unterschied könnte zufällig sein.")
```
