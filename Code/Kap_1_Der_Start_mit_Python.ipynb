```python
# Einige kleine Rechenoperationen (wie mit dem Taschenrechner)
x = 5+7
print(f"Wert für x: {x}")
y=6*4
print(f"Wert für y: {y}")
z=x+y
print(f"Wert für z: {z}")
```

```python
# Die Grundrechenoperationen
Addieren = 5+7
Subtrahieren = 9-6
Multiplizieren = 5*6
Dividieren = 12/5
Potenzieren = 3**2
print(Addieren, Subtrahieren, Multiplizieren, Dividieren, Potenzieren)

# Mit schönem Output
print(f"Ergebnis der Addition: {Addieren}")
print(f"Ergebnis der Subtraktion: {Subtrahieren}")
print(f"Ergebnis der Multiplikation: {Multiplizieren}")
print(f"Ergebnis der Division: {Dividieren}")
print(f"Ergebnis der Potenzrechnung: {Potenzieren}")
```

```python
wurzel_2 = 9**(1/2)
print(f"Quadratwurzel: {wurzel_2}")

wurzel_3 = 8**(1/3)
print(f"Dritte Wurzel: {wurzel_3}")
```

## 1.6 Bibliotheken in Python

```python
# Zuerst laden wir die Bibliothek, um ihre Werkzeuge zu nutzen
import numpy as np

# Wir erstellen ein spezielles numpy-Array, das ist wie eine verbesserte Liste
meine_zahlen_np = np.array([4, 9, 16, 25])

# Und jetzt nutzen wir das leistungsfähige Werkzeug von numpy
# um die Wurzeln zu ziehen
wurzeln_np = np.sqrt(meine_zahlen_np)

print(wurzeln_np)
# Ausgabe: [2. 3. 4. 5.]
```
