```python
import matplotlib.pyplot as plt
bins=[0,100,200,300,400,500,600,700] #Klassengrenzen
plt.hist(df3['Warmmiete'], bins=bins,  edgecolor='black')
plt.xlabel('Warmmiete')
plt.ylabel('Häufigkeit')
plt.title('Histogramm mit Klassenbreiten von 100')
plt.show()
```

```python
# Häufigkeit jeder Kategorie zählen
Geschlecht_counts = df3['Geschlecht'].value_counts()

# Kreisdiagramm erstellen
plt.pie(Geschlecht_counts, labels=Geschlecht_counts.index, autopct='%1.1f%%')

# Titel hinzufügen
plt.title('Verteilung des Geschlechts')

# Diagramm anzeigen
plt.show()
```

```python
# Häufigkeit jeder Kategorie zählen
kategorie_counts = df3['Namensherkunft'].value_counts()

# Balkendiagramm erstellen
plt.barh(kategorie_counts.index, kategorie_counts.values)

# Achsenbeschriftungen und Titel hinzufügen
plt.xlabel('Namensherkunft')
plt.ylabel('Anzahl')
plt.title('Häufigkeiten der Kategorien')

# Diagramm anzeigen
plt.show()
```

```python
# Scatterplot erstellen
plt.scatter(df3['Sport'], df3['Warmmiete'])

# Achsenbeschriftungen und Titel hinzufügen
plt.xlabel('Sportaktivitäten pro Woche')
plt.ylabel('Monatliche Warmmiete')
plt.title('Scatterplot von Sportaktivitäten vs. Warmmiete')

# Diagramm anzeigen
plt.show()
```
