## 5.1 Binomialverteilung

```python
from scipy.stats import binom

# Parameter der Binomialverteilung
n = 20  # Stichprobengröße
theta = 0.05 # Wahrscheinlichkeit für einen Defekt

# 1. Wahrscheinlichkeit für genau zwei defekte Smartphones
wahrscheinlichkeit_genau_zwei = binom.pmf(2, n, theta)
print(f"Wahrscheinlichkeit für genau zwei defekte Smartphones: {wahrscheinlichkeit_genau_zwei:.4f}")
```

```python
# 2. Wahrscheinlichkeit für höchstens ein defektes Smartphone (0 oder 1)
wahrscheinlichkeit_hoechstens_eins = binom.cdf(1, n, theta)
print(f"Wahrscheinlichkeit für höchstens ein defektes Smartphone: {wahrscheinlichkeit_hoechstens_eins:.4f}")
```

```python
# 3. Wahrscheinlichkeit für mindestens drei defekte Smartphones (3, 4, ..., 20)
wahrscheinlichkeit_mindestens_drei = 1 - binom.cdf(2, n, theta)
print(f"Wahrscheinlichkeit für mindestens drei defekte Smartphones: {wahrscheinlichkeit_mindestens_drei:.4f}")
```

```python
# 4. Wahrscheinlichkeit für mehr als vier defekte Smartphones (5, 6, ..., 20)
wahrscheinlichkeit_mehr_als_vier = 1 - binom.cdf(4, n, theta)
print(f"Wahrscheinlichkeit für mehr als vier defekte Smartphones: {wahrscheinlichkeit_mehr_als_vier:.4f}")
```

```python
# 5. Erwartungswert und Standardabweichung
erwartungswert = binom.mean(n, theta)
standardabweichung = binom.std(n, theta)
print(f"\nErwartungswert der Anzahl defekter Smartphones: {erwartungswert:.2f}")
print(f"Standardabweichung der Anzahl defekter Smartphones: {standardabweichung:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_werte = np.arange(0, n + 1)  # Alle möglichen Werte für x
pmf_werte = binom.pmf(x_werte, n, theta)  # pmf für jeden x-Wert

plt.bar(x_werte, pmf_werte)
plt.xlabel("x")
plt.ylabel("W(X=x)")
plt.title("Wahrscheinlichkeitsfunktion (pmf)")
plt.show()
```

```python
x_werte = np.arange(0, n + 1)  # Alle möglichen Werte für x
cdf_werte = binom.cdf(x_werte, n, theta)  # cdf für jeden x-Wert

plt.bar(x_werte, cdf_werte)
plt.xlabel("x")
plt.ylabel("W(X<=x)")
plt.title("Verteilungsfunktion (cdf)")
plt.show()
```

## 5.2 Hypergeometrische Verteilung

```python
import numpy as np
from scipy.stats import hypergeom
import matplotlib.pyplot as plt

# Parameter der hypergeometrischen Verteilung
N = 50  # Gesamtanzahl der Mitarbeiter in der Population
M = 12  # Anzahl der Mitarbeiter mit Online-Marketing-Weiterbildung in der Population
n = 8   # Anzahl der zufällig ausgewählten Mitarbeiter

# 1. Wahrscheinlichkeit für genau 3 Mitarbeiter mit Weiterbildung (pmf)
x_1 = 3
pmf_1 = hypergeom.pmf(x_1, N, M, n)
print(f"1. Wahrscheinlichkeit für genau {x_1} Mitarbeiter mit Weiterbildung: {pmf_1:.4f}")
```

```python
# 2. Wahrscheinlichkeit für höchstens 2 Mitarbeiter mit Weiterbildung (cdf)
x_2 = 2
cdf_2 = hypergeom.cdf(x_2, N, M, n)
print(f"2. Wahrscheinlichkeit für höchstens {x_2} Mitarbeiter mit Weiterbildung: {cdf_2:.4f}")
```

```python
# 3. Wahrscheinlichkeit für mindestens 4 Mitarbeiter mit Weiterbildung (1 - cdf)
x_3_kleiner = 3  # Wir berechnen die Wahrscheinlichkeit für höchstens 3 und ziehen sie von 1 ab
cdf_3 = hypergeom.cdf(x_3_kleiner, N, M, n)
W_3 = 1 - cdf_3
print(f"3. Wahrscheinlichkeit für mindestens 4 Mitarbeiter mit Weiterbildung: {W_3:.4f}")
```

```python
# Visualisierung der Wahrscheinlichkeitsfunktion
x = np.arange(0, n + 1)
pmf_werte = hypergeom.pmf(x, N, M, n)

plt.figure(figsize=(8, 6))
plt.bar(x, pmf_werte, color='skyblue')
plt.xlabel('Anzahl der Mitarbeiter mit Online-Marketing-Weiterbildung')
plt.ylabel('Wahrscheinlichkeit')
plt.title('Hypergeometrische Verteilung der ausgewählten Mitarbeiter')
plt.xticks(x)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Visualisierung der Verteilungsfunktion
x = np.arange(0, n + 1)
cdf_werte = hypergeom.cdf(x, N, M, n)

plt.figure(figsize=(8, 6))
plt.bar(x, cdf_werte, color='skyblue')
plt.xlabel('Anzahl der Mitarbeiter mit Online-Marketing-Weiterbildung')
plt.ylabel('Wahrscheinlichkeit')
plt.title('Hypergeometrische Verteilung der höchstens ausgewählten Mitarbeiter')
plt.xticks(x)
plt.grid(axis='y', linestyle='--')
plt.show()
```

## 5.3 Normalverteilung

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Parameter
mittelwert = 0.025
standardabweichung = 0.005

# Definieren einen Bereich von Klickraten für die grafische Darstellung
x = np.linspace(mittelwert - 3 * standardabweichung, mittelwert + 3 * standardabweichung, 200)
pdf = st.norm.pdf(x, mittelwert, standardabweichung)

# Allgemeine Funktion zum Plotten
def plot_normal_distribution(title, x_values, pdf_values, shaded_area=None, shaded_label=None, vertical_line=None, vertical_label=None):
    plt.figure(figsize=(10, 6))
    plt.plot(x_values * 100, pdf_values, 'b-', label=f'$\\mu$={mittelwert*100:.1f}%, $\\sigma$={standardabweichung*100:.2f}%')
    if shaded_area is not None:
        plt.fill_between(x_values * 100, 0, pdf_values, where=shaded_area, color='orange', alpha=0.5, label=shaded_label)
    if vertical_line is not None:
        plt.axvline(vertical_line * 100, color='red', linestyle='--', label=vertical_label)
    plt.title(title)
    plt.xlabel('Klickrate (%)')
    plt.ylabel('Wahrscheinlichkeitsdichte')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
untere_grenze = 0.02
obere_grenze = 0.03
wahrscheinlichkeit_bereich = st.norm.cdf(obere_grenze, mittelwert, standardabweichung) - st.norm.cdf(untere_grenze, mittelwert, standardabweichung)
print(f"Die Wahrscheinlichkeit einer Klickrate zwischen {untere_grenze*100}% und {obere_grenze*100}% beträgt: {wahrscheinlichkeit_bereich:.4f}")

# Grafische Darstellung
bereich_auswahl = (x >= untere_grenze) & (x <= obere_grenze)
plot_normal_distribution(
    "Wahrscheinlichkeit einer Klickrate zwischen 2% und 3%",
    x,
    pdf,
    shaded_area=bereich_auswahl,
    shaded_label=f'P({untere_grenze*100:.1f}% $\\leq$ CTR $\\leq$ {obere_grenze*100:.1f}%) = {wahrscheinlichkeit_bereich:.4f}'
)
```

```python
grenzwert_oben = 0.035
wahrscheinlichkeit_oberhalb = 1 - st.norm.cdf(grenzwert_oben, mittelwert, standardabweichung)
print(f"Die Wahrscheinlichkeit einer Klickrate über {grenzwert_oben*100:.1f}% beträgt: {wahrscheinlichkeit_oberhalb:.4f}")

# Grafische Darstellung
oberhalb_auswahl = (x > grenzwert_oben)
plot_normal_distribution(
    "Wahrscheinlichkeit einer Klickrate über 3.5%",
    x,
    pdf,
    shaded_area=oberhalb_auswahl,
    shaded_label=f'P(CTR > {grenzwert_oben*100:.1f}%) = {wahrscheinlichkeit_oberhalb:.4f}',
    vertical_line=grenzwert_oben,
    vertical_label=f'CTR = {grenzwert_oben*100:.1f}%'
)
```

```python
prozentrang_oben = 0.90  # Wir suchen den Wert, bei dem 90% der Daten darunter liegen
schwellenwert_oben = st.norm.ppf(prozentrang_oben, mittelwert, standardabweichung)
print(f"Die Klickrate, oberhalb der die besten 10% der Nutzer liegen, beträgt: {schwellenwert_oben:.4f}")

# Grafische Darstellung
plot_normal_distribution(
    "Schwellenwert für die oberen 10% der Klickraten",
    x,
    pdf,
    vertical_line=schwellenwert_oben,
    vertical_label=f'Top 10% ab CTR = {schwellenwert_oben*100:.2f}%'
)
```

## 5.4 Von der Normalverteilung zur Standardnormalverteilung

```python
import scipy.stats as st
import numpy as np

# Parameter für Kampagne A
mittelwert_a = 0.05
standardabweichung_a = 0.01
ziel_conversion_a = 0.07

# Parameter für Kampagne B
mittelwert_b = 0.08
standardabweichung_b = 0.02
ziel_conversion_b = 0.11

# 1. z-Wert Berechnung
z_wert_a = (ziel_conversion_a - mittelwert_a) / standardabweichung_a
z_wert_b = (ziel_conversion_b - mittelwert_b) / standardabweichung_b

print(f"z-Wert für eine Conversion-Rate von {ziel_conversion_a*100:.1f}% in Kampagne A: {z_wert_a:.2f}")
print(f"z-Wert für eine Conversion-Rate von {ziel_conversion_b*100:.1f}% in Kampagne B: {z_wert_b:.2f}")

# 2. Wahrscheinlichkeit basierend auf z-Werten (Standardnormalverteilung)
# Die Wahrscheinlichkeit P(X >= x) für eine Normalverteilung entspricht P(Z >= z) für die Standardnormalverteilung,
# wobei z der z-Wert von x ist. Wir verwenden 1 - cdf(z), um die Wahrscheinlichkeit "größer oder gleich" zu erhalten.

wahrscheinlichkeit_a = 1 - st.norm.cdf(z_wert_a)
wahrscheinlichkeit_b = 1 - st.norm.cdf(z_wert_b)

print(f"Wahrscheinlichkeit für eine Conversion-Rate von mindestens {ziel_conversion_a*100:.1f}% in Kampagne A: {wahrscheinlichkeit_a:.4f}")
print(f"Wahrscheinlichkeit für eine Conversion-Rate von mindestens {ziel_conversion_b*100:.1f}% in Kampagne B: {wahrscheinlichkeit_b:.4f}")

# 3. Relativer Vergleich
print("\nRelativer Vergleich:")
if wahrscheinlichkeit_a > wahrscheinlichkeit_b:
    print("Relativ gesehen zeigt Kampagne A eine höhere Wahrscheinlichkeit, Nutzer mit der Ziel-Conversion-Rate oder höher zu erreichen.")
elif wahrscheinlichkeit_b > wahrscheinlichkeit_a:
    print("Relativ gesehen zeigt Kampagne B eine höhere Wahrscheinlichkeit, Nutzer mit der Ziel-Conversion-Rate oder höher zu erreichen.")
else:
    print("Relativ gesehen zeigen beide Kampagnen eine ähnliche Wahrscheinlichkeit, Nutzer mit der Ziel-Conversion-Rate oder höher zu erreichen.")
```

## 5.5 Demonstration des zentralen Grenzwertsatzes

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Gleichverteilte Datenmenge erstellen
Daten = np.random.uniform(0, 1, 1000)

# Stichprobenumfänge definieren
Stichproben_Umfaenge = [10, 20, 30]

# Anzahl der Stichproben pro Umfang
Anzahl_Stichproben = 1000

# Mittelwerte der Stichproben berechnen
Stichprobenmittelwerte = []
for size in Stichproben_Umfaenge:
    means = [np.mean(np.random.choice(Daten, size=size)) for _ in range(Anzahl_Stichproben)]
    Stichprobenmittelwerte.append(means)

# Histogramme erstellen
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Histogramm für die gleichverteilte Datenmenge
sns.histplot(Daten, ax=axes[0, 0], kde=False)
axes[0, 0].set_title("Gleichverteilte Datenmenge")
axes[0, 0].set_xlabel("Wert")
axes[0, 0].set_ylabel("Häufigkeit")

# Histogramme der Stichprobenmittelwerte
for i, size in enumerate(Stichproben_Umfaenge):
    sns.histplot(Stichprobenmittelwerte[i], ax=axes[1, i], kde=True)
    axes[1, i].set_title(f"Stichprobenumfang: {size}")
    axes[1, i].set_xlabel("Stichprobenmittelwert")
    axes[1, i].set_ylabel("Häufigkeit")

# Leere Subplots entfernen
fig.delaxes(axes[0,1])
fig.delaxes(axes[0,2])

plt.tight_layout()
plt.show()
```

## 5.6 Chi-Quadrat-Verteilung

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# 1. Definiere die Freiheitsgrade (df)
dfs = [2, 3, 5, 10]

# 2. Definiere den Bereich für die x-Achse
# Chi-Quadrat-Werte sind immer >= 0
x = np.linspace(0, 25, 500)

# 3. Initialisiere den Plot
plt.figure(figsize=(10, 6))
plt.title('Chi-Quadrat-Verteilung für verschiedene Freiheitsgrade')
plt.xlabel('Chi-Quadrat-Wert')
plt.ylabel('Wahrscheinlichkeitsdichte (PDF)')

# 4. Plotten jeder Verteilung
for df in dfs:
    # Berechne die Wahrscheinlichkeitsdichte (PDF)
    pdf = chi2.pdf(x, df)

    # Plotten mit einer Beschriftung
    plt.plot(x, pdf, label=f'df = {df}')

# 5. Füge Legende und Raster hinzu
plt.legend(title='Freiheitsgrade')
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(0, 0.6) # Setzt eine sinnvolle Obergrenze
plt.xlim(0, 25)  # Setzt eine sinnvolle Obergrenze

# 6. Zeige den Plot an
plt.show()
```

## 5.7 Student-t-Verteilung

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# 1. Definiere die Freiheitsgrade (df)
dfs = [2, 4, 10, 30]

# 2. Definiere den Bereich für die x-Achse
# Die t-Verteilung ist symmetrisch um 0 und erstreckt sich von -inf bis +inf
x = np.linspace(-5, 5, 500)

# 3. Initialisiere den Plot
plt.figure(figsize=(10, 6))
plt.title('Student-$t$-Verteilung für verschiedene Freiheitsgrade')
plt.xlabel('t-Wert')
plt.ylabel('Wahrscheinlichkeitsdichte (PDF)')

# 4. Plotten jeder t-Verteilung
for df in dfs:
    # Berechne die Wahrscheinlichkeitsdichte (PDF)
    pdf_t = t.pdf(x, df)

    # Plotten mit einer Beschriftung
    plt.plot(x, pdf_t, label=f't, df = {df}')

# 5. Plotten der Standardnormalverteilung (als Referenz)
pdf_norm = norm.pdf(x, 0, 1) # Mittelwert=0, Standardabweichung=1
plt.plot(x, pdf_norm, label='Normalverteilung', linestyle='--', color='black')

# 6. Füge Legende und Raster hinzu
plt.legend(title='Verteilungen')
plt.grid(True, linestyle='--', alpha=0.6)

# 7. Zeige den Plot an
plt.show()
```
