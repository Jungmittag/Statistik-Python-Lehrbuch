### 4.1.4 Geometrisches Mittel

```python
dax=pd.read_excel('Data/DAX_Renditen.xlsx', index_col='Datum', parse_dates=['Datum'])
print(dax.head(8))
```

```python
import matplotlib.pyplot as plt

dax['DAX_Rendite'].plot()
plt.show()
```

```python
dax['DAX_Wachstumsfaktor'] = dax['DAX_Rendite'] / 100 + 1
print(dax.head(8))
```

```python
import numpy as np

geometrisches_mittel_1 = np.exp(np.mean(np.log(dax['DAX_Wachstumsfaktor'])))
print(f"Geometrisches Mittel: {geometrisches_mittel_1:.5f}")

rendite_geomittel_1=(geometrisches_mittel_1-1)*100
print(f"Durchschnittliche monatliche Rendite des DAX: {rendite_geomittel_1:.3f}")
```

```python
from scipy.stats import gmean

geometrisches_mittel_2 = gmean(dax['DAX_Wachstumsfaktor'])
print(f"Geometrisches Mittel: {geometrisches_mittel_2:.5f}")
rendite_geomittel_2=(geometrisches_mittel_2-1)*100
print(f"Durchschnittliche monatliche Rendite des DAX: {rendite_geomittel_2:.3f}")
```

### 4.1.5 Harmonisches Mittel

```python
def durchschnittliche_geschwindigkeit(strecken, geschwindigkeiten):
    """Berechnet die durchschnittliche Geschwindigkeit für eine gegebene Strecke."""

    if len(strecken) != len(geschwindigkeiten):
        raise ValueError("Die Anzahl der Strecken und Geschwindigkeiten muss übereinstimmen.")

    gesamtstrecke = sum(strecken)
    gesamtzeit = sum(strecke / geschwindigkeit for strecke, geschwindigkeit in zip(strecken, geschwindigkeiten))

    if gesamtzeit == 0:
        return 0  # Vermeidet Division durch Null

    return gesamtstrecke / gesamtzeit

# Gegebene Daten
strecken = [105, 120, 35]  # Strecken in km
geschwindigkeiten = [70, 60, 35]  # Geschwindigkeiten in km/h

# Berechnung der durchschnittlichen Geschwindigkeit
durchschnitt = durchschnittliche_geschwindigkeit(strecken, geschwindigkeiten)

# Ausgabe des Ergebnisses
print(f"Die durchschnittliche Geschwindigkeit beträgt: {durchschnitt:.2f} km/h")
```

## 4.2 Streuungsmaße

### 4.2.1 Spannweite

```python
spannweite_sport = df['Sport'].max() - df['Sport'].min()
print(f"Die Spannweite der Variable Sport beträgt: {spannweite_sport}")

spannweite_miete = df['Warmmiete'].max() - df['Warmmiete'].min()
print(f"Die Spannweite der Variable Warmmiete beträgt: {spannweite_miete}")
```

### 4.2.2 Mittlere absolute Abweichung

```python
import numpy as np

median_sport = df['Sport'].median()
mad_median_sport = np.mean(np.abs(df['Sport'] - median_sport))
print(f"Mittlere absolute Abweichung (Median) für die Variable Sport: {mad_median_sport}")

median_miete = df['Warmmiete'].median()
mad_median_miete = np.mean(np.abs(df['Warmmiete'] - median_miete))
print(f"Mittlere absolute Abweichung (Median) für die Variable Warmmiete: {mad_median_miete}")
```

```python
mean_sport = df['Sport'].mean()
mad_mean_sport = np.mean(np.abs(df['Sport'] - mean_sport))
print(f"Mittlere absolute Abweichung (arithmetisches Mittel) für die Variable Sport: {mad_mean_sport:.2f}")

mean_miete = df['Warmmiete'].mean()
mad_mean_miete = np.mean(np.abs(df['Warmmiete'] - mean_miete))
print(f"Mittlere absolute Abweichung (arithmetisches Mittel) für die Variable Warmmiete: {mad_mean_miete:.2f}")
```

### 4.2.3 Varianz und Standardabweichung

```python
varianz_sport = df['Sport'].var(ddof=0) # ddof=0  führt zur Berechnung der Varianz für eine Grundgesamtheit
print(f"Varianz für die Variable Sport: {varianz_sport:.2f}")

stdabw_sport = df['Sport'].std(ddof=0) # ddof=0  führt zur Berechnung der Standardabweichung für eine Grundgesamtheit
print(f"Standardabweichung für die Variable Sport: {stdabw_sport:.2f}")
```

```python
varianz_miete = df['Warmmiete'].var(ddof=0) # ddof=0  führt zur Berechnung der Varianz für eine Grundgesamtheit
print(f"Varianz für die Variable Warmmiete: {varianz_miete:.2f}")

stdabw_miete = df['Warmmiete'].std(ddof=0) # ddof=0  führt zur Berechnung der Standardabweichung für eine Grundgesamtheit
print(f"Standardabweichung für die Variable Warmmiete: {stdabw_miete:.2f}")
```

### 4.2.4 Variationskoeffizient

```python
import numpy as np
Aktie_A = [1.23,1.50,2.27,0.54,2.40,3.10,5.18]
mu_A = np.mean(Aktie_A)
sigma_A =np.std(Aktie_A,ddof=0)
VK_A=sigma_A/mu_A*100
print('Unternehmen A:')
print('Mittelwert = {:.2f}; Standardabweichung = {:.2f}; VK = {:.2f}'.format(mu_A,sigma_A,VK_A))

Aktie_B = [171,169,182,159,164,175,180,173,176,177]
mu_B = np.mean(Aktie_B)
sigma_B =np.std(Aktie_B,ddof=0)
VK_B=sigma_B/mu_B*100
print('Unternehmen B:')
print('Mittelwert = {:.2f}; Standardabweichung = {:.2f}; VK = {:.2f}'.format(mu_B,sigma_B,VK_B))
```
