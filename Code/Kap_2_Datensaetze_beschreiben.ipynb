## 2.1 Häufigkeitsverteilungen für Einzelwerte

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sport=[4, 1, 4, 3, 1, 0, 4, 2, 2, 1, 2, 3, 3, 3, 2, 2, 4, 0, 4, 1, 1, 0, 2, 1, 2,1, 0, 3, 2, 2, 1, 2, 2, 3, 1, 2, 1, 2, 3, 4, 3, 4, 1, 5, 1, 3, 3, 3, 2, 1]
print(sport)
```

```python
Werte, Anzahl = np.unique(sport, return_counts=True)
print(Werte, Anzahl)

# Mit schöner Ausgabe
print(f"Die Merkmalsausprägungen sind: {Werte}")
print(f"Ihre Häufigkeiten sind: {Anzahl}")
```

```python
N=len(sport)
print(f"N = {N}")
```

```python
rel_freqs=Anzahl/N
print(f"Die Merkmalsausprägungen sind: {Werte}")
print(f"Ihre relativen Häufigkeiten sind: {rel_freqs}")
```

```python
print("Relative Häufigkeiten:")
for i, val in enumerate(Werte):
    print(val, rel_freqs[i])
```

```python
plt.stem(Werte, rel_freqs)
plt.xlabel("Sport pro Woche")
plt.ylabel("Relative Häufigkeiten")
plt.title("Stabdiagramm");
```

```python
sum_freq=np.cumsum(Anzahl)
print(f"Die Merkmalsausprägungen sind: {Werte}")
print(f"Ihre Summenhäufigkeiten sind: {sum_freq}")
```

```python
rel_sum_freq=np.cumsum(rel_freqs)
print(f"Die Merkmalsausprägungen sind: {Werte}")
print(f"Ihre relativen Summenhäufigkeiten sind: {rel_sum_freq}")
```

```python
plt.step(Werte, sum_freq)
plt.xlabel("Sport pro Woche")
plt.ylabel("Summenhäufigkeiten")
plt.title("Stufendiagramm für die absoluten Häufigkeiten");
```

```python
plt.step(Werte, rel_sum_freq)
plt.xlabel("Sport pro Woche")
plt.ylabel("Relative Summenhäufigkeiten")
plt.title("Stufendiagramm für die relativen Häufigkeiten");
```

```python
df=pd.DataFrame(sport)
freq=df.value_counts()
print(freq)
```

```python
freq_sorted=df.value_counts().sort_index()
print(freq_sorted)
```

```python
freq_rel=df.value_counts(normalize=True).sort_index()
print(freq_rel)
```

## 2.2 Häufigkeitsverteilungen bei klassifizierten Daten

```python
Miete=[698, 418, 178, 0, 474, 0, 353, 334, 146, 260, 353, 245, 405, 508, 248, 320, 371, 251, 210, 0, 273, 298, 457, 602, 225, 305, 273, 390, 216, 467, 228, 261, 372, 485, 234, 263, 591, 311, 385, 275, 370, 359, 471, 366, 313, 425, 180, 242, 140, 581]
print(Miete)
```

```python
N_1=len(Miete)
print(f"N_1 = {N_1}")
```

```python
Miete_klass=pd.cut(Miete,7)
Miete_klass.value_counts()
```

```python
bins=[0,100,200,300,400,500,600,700] #Klassengrenzen
labels=['0-100', 'über 100-200', 'über 200-300', 'über 300-400', 'über 400-500', 'über 500-600','über 600-700']
df_1=pd.DataFrame({'Werte':Miete})
df_1['Klasse'] = pd.cut(df_1['Werte'], bins=bins, labels=labels, right=False)
häufigkeiten = df_1['Klasse'].value_counts().sort_index()
print(häufigkeiten)
```

```python
rel_häufigkeiten = df_1['Klasse'].value_counts(normalize=True).sort_index()
print(rel_häufigkeiten)
```

```python
plt.hist(Miete, bins=bins, edgecolor='black')
plt.xlabel('Warmmiete')
plt.ylabel('Häufigkeit')
plt.title('Histogramm mit Klassenbreiten von 100')
plt.show()
```

```python
Klassenbreiten=[0,100,200,250,300,350,400,500,600,700]
plt.hist(Miete, bins=Klassenbreiten, edgecolor='black', color='lightgreen')
plt.xlabel('Warmmiete')
plt.ylabel('Häufigkeit pro Einheit der Klassenbreite')
plt.title('Histogramm mit unterschiedlichen Klassenbreiten')
plt.show()
```

```python
Klassenbreiten=[0,100,200,250,300,350,400,500,600,700]
plt.hist(Miete, bins=Klassenbreiten, edgecolor='black', color='lightyellow',density=True)
plt.xlabel('Warmmiete')
plt.ylabel('Dichte')
plt.title('Histogramm mit unterschiedlichen Klassenbreiten')
plt.show()
```

```python
sum_häufigkeiten=np.cumsum(häufigkeiten)
neue_labels={'0-100':'bis 100','über 100-200':'bis 200','über 200-300':'bis 300','über 300-400':'bis 400','über 400-500':'bis 500','über 500-600':'bis 600','über 600-700':'bis 700'}
sum_häufigkeiten=sum_häufigkeiten.rename(index=neue_labels)
print(sum_häufigkeiten)
```

```python
sum_rel_häufigkeiten=np.cumsum(rel_häufigkeiten)
neue_labels={'0-100':'bis 100','über 100-200':'bis 200','über 200-300':'bis 300','über 300-400':'bis 400','über 400-500':'bis 500','über 500-600':'bis 600','über 600-700':'bis 700'}
sum_rel_häufigkeiten=sum_rel_häufigkeiten.rename(index=neue_labels)
print(sum_rel_häufigkeiten)
```

```python
plt.plot(sum_rel_häufigkeiten.index, sum_rel_häufigkeiten.values, marker='o')
plt.xlabel('Warmmieten')
plt.ylabel('Relative Summenhäufigkeit')
plt.title('Summenpolygon der klassifizierten Daten')
plt.grid(True)  # Optional: Gitternetzlinien hinzufügen
plt.show()
```
