## 6.1 Konfidenzintervalle für den Anteilswert

```python
import numpy as np
from scipy import stats

# Daten
n = 500  # Stichprobengröße
x = 300  # Anzahl der Erfolge
p = x / n # Stichprobenanteil
print(f"Anteilswert in der Stichprobe: {p:.2f}")

# Konfidenzniveau
konfidenzniveau = 0.95
alpha = 1 - konfidenzniveau

# Kritischer Z-Wert
z_kritisch = stats.norm.ppf(1 - alpha / 2)
print(f"Kritischer Z-Wert: {z_kritisch:.2f}")

# Standardfehler
se_p = np.sqrt((p * (1 - p)) / n)
print(f"Standardfehler: {se_p:.4f}")

# Fehlerspanne 
fehlerspanne_theta = z_kritisch * se_p
print(f"Fehlerspanne: {fehlerspanne_theta:.4f}")

# Konfidenzintervall
untere_grenze_theta = p - fehlerspanne_theta
obere_grenze_theta = p + fehlerspanne_theta
print(f"Konfidenzintervall (manuell): [{untere_grenze_theta:.4f}, {obere_grenze_theta:.4f}]")
```

```python
from statsmodels.stats.proportion import proportion_confint

Konfidenzintervall_p = proportion_confint(count=x, nobs=n, alpha=alpha, method='normal') # x, n und alpha werden aus dem vorherigen Code übernommen
print(f"Konfidenzintervall (statsmodels): [{Konfidenzintervall_p[0]:.4f}, {Konfidenzintervall_p[1]:.4f}]")
```

## 6.2 Konfidenzintervalle für das arithmetische Mittel

```python
import numpy as np
from scipy import stats

# Daten
x_quer = 22.5 # Stichprobenmittelwert
s = 2.1   # Stichprobenstandardabweichung
n_mw = 30        # Stichprobengröße
freiheitsgrade = n_mw - 1 # Freiheitsgrade

# Konfidenzniveau
konfidenzniveau_mw = 0.90
alpha_mw = 1 - konfidenzniveau_mw

# Kritischer T-Wert
t_kritisch = stats.t.ppf(1 - alpha_mw / 2, freiheitsgrade)
print(f"Kritischer T-Wert: {t_kritisch:.2f}")

# Standardfehler des Mittelwerts
s_x_quer = s / np.sqrt(n_mw)
print(f"Standardfehler des Mittelwerts: {s_x_quer:.4f}")

# Fehlerspanne
fehlerspanne_mw = t_kritisch * s_x_quer
print(f"Fehlerspanne: {fehlerspanne_mw:.4f}")

# Konfidenzintervall
untere_grenze_mw = x_quer - fehlerspanne_mw
obere_grenze_mw = x_quer + fehlerspanne_mw
print(f"Konfidenzintervall (manuell): [{untere_grenze_mw:.4f}, {obere_grenze_mw:.4f}]")
```

```python
conf_interval_scipy_mean = stats.t.interval(confidence=konfidenzniveau_mw,
                                        df=freiheitsgrade,
                                        loc=x_quer,
                                        scale=s_x_quer)
print(f"Konfidenzintervall (scipy.stats): [{conf_interval_scipy_mean[0]:.4f}, {conf_interval_scipy_mean[1]:.4f}]")
```

## 6.3 Konfidenzintervalle für die Varianz

```python
import numpy as np
from scipy.stats import chi2

# Die Stichprobendaten der Stückkosten in Euro
kosten = np.array([
    10.25, 9.80, 10.00, 10.15, 9.90, 10.30, 10.05, 10.10, 9.75,
    10.40, 9.95, 10.30, 10.00, 10.10, 10.25, 9.90, 10.00, 10.15,
    10.05, 10.20, 9.90, 10.35, 9.85, 10.10, 10.20
])

# Parameter
n = len(kosten)           # Stichprobengröße
df = n - 1                # Freiheitsgrade
konfidenzniveau = 0.95    # 95%
alpha = 1 - konfidenzniveau

# 1. Stichprobenvarianz (s^2) berechnen
# Die Funktion np.var() mit ddof=1 berechnet die Stichprobenvarianz s^2 mit n-1
s_quadrat = np.var(kosten, ddof=1)


print(f"Stichprobengröße (n): {n}")
print(f"Freiheitsgrade (df): {df}")
print(f"Stichprobenvarianz (s^2): {s_quadrat:.6f} Euro^2")
```

```python
# Die Chi-Quadrat-Werte (Quantile) für die untere und obere Grenze
# ppf = Percent Point Function (Inverse der CDF)
chi2_untere_grenze = chi2.ppf(alpha / 2, df)     # Chi^2_0.025
chi2_obere_grenze = chi2.ppf(1 - alpha / 2, df) # Chi^2_0.975

print(f"Chi^2-Wert für 0.025: {chi2_untere_grenze:.4f}")
print(f"Chi^2-Wert für 0.975: {chi2_obere_grenze:.4f}")
```

```python
# Untere Intervallgrenze
untere_grenze = (df * s_quadrat) / chi2_obere_grenze

# Obere Intervallgrenze
obere_grenze = (df * s_quadrat) / chi2_untere_grenze

# Ausgabe des Ergebnisses
print("\n--- Ergebnis ---")
print(f"95%-Konfidenzintervall für die Varianz (sigma^2): [{untere_grenze:.4f}, {obere_grenze:.4f}] Euro^2")

# Optional: Konfidenzintervall für die Standardabweichung (sigma)
ci_std_dev = [np.sqrt(untere_grenze), np.sqrt(obere_grenze)]
print(f"95%-Konfidenzintervall für die Standardabweichung (sigma): [{ci_std_dev[0]:.4f}, {ci_std_dev[1]:.4f}] Euro")
```
