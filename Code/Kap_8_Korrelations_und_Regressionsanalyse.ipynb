# Kapitel 8: Korrelations- und Regressionsanalyse

```python
import pandas as pd

df_note=pd.read_excel('Data/uni_note.xlsx')
print(df_note.head(4))
```

```python
print(df_note.describe().round(2))
```

## 8.1 Korrelationsanalyse

```python
# Korrelation zwischen 'Notendurchschnitt im Studium' und 'Abitursnote'
Korrelation_Uni_Abi_Note = df_note['colGPA'].corr(df_note['hsGPA'])
print(f"Korrelation zwischen Notendurchschnitt im Studium und Abiturnote: {Korrelation_Uni_Abi_Note:.4f}")
```

```python
import matplotlib.pyplot as plt

# Scatterplot erstellen
plt.scatter(df_note['hsGPA'], df_note['colGPA'])

# Achsenbeschriftungen und Titel hinzufügen
plt.xlabel('Abiturnote')
plt.ylabel('Notendurchschnitt im Studium')
plt.title('Scatterplot von Abiturnote vs. Durchschnittsnote im Studium')

# Diagramm anzeigen
plt.show()
```

```python
# Korrelationsmatrix berechnen
Korrelationsmatrix = df_note.corr().round(3)

print("Korrelationsmatrix für alle numerischen Variablen:")
print(Korrelationsmatrix)
```

## 8.2 Regressionsanalyse

```python
import statsmodels.formula.api as smf

# KQ-Regression
reg_1 = smf.ols(formula="colGPA ~ hsGPA", data=df_note)
results_1 = reg_1.fit()

# Ausgabe der Ergebnisse mit summary
print(f"results_1.summary(): \n{results_1.summary()}\n")
```

```python
# Koeffizienten extrahieren
Konstante = results_1.params['Intercept']
Steigung = results_1.params['hsGPA']

# 2. Scatterplot erstellen
# X-Achse: hsGPA (Abitursnote), Y-Achse: colGPA (Studiennote)
plt.scatter(df_note['hsGPA'], df_note['colGPA'], label='Datenpunkte')

# 3. Regressionsgerade hinzufügen
# Vorhersagewerte berechnen: y = Achsenabschnitt (Konstante) + Steigung * x
# plt.plot(X-Werte, Y-Werte)
plt.plot(df_note['hsGPA'], Konstante + Steigung * df_note['hsGPA'], color='red',
         label=f'KQ-Regressionsgerade (y={Konstante:.2f} + {Steigung:.2f}x)')

# Achsenbeschriftungen und Titel hinzufügen
plt.xlabel('Abiturnote')
plt.ylabel('Notendurchschnitt im Studium')
plt.title('KQ-Regression: Abiturnote vs. Durchschnittsnote im Studium')

# Legende und Diagramm anzeigen
plt.legend()
plt.show()
```

```python
reg_2 = smf.ols(formula="colGPA ~ ACT", data=df_note)
results_2 = reg_2.fit()
print(f"results_2.summary(): \n{results_2.summary()}\n")
```

```python
reg_3 = smf.ols(formula="colGPA ~ hsGPA + ACT", data=df_note)
results_3 = reg_3.fit()
print(f"results_3.summary(): \n{results_3.summary()}\n")
```

```python
reg_4 = smf.ols(formula="colGPA ~ hsGPA + male", data=df_note)
results_4 = reg_4.fit()
print(f"results_4.summary(): \n{results_4.summary()}\n")
```
